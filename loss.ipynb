```python
import matplotlib.pyplot as plt
import re
```

```python
# log_file_path = 'cgan_small_step.log'  
log_file_path = 'transformer.log'  

# d_losses = []
# g_losses = []
    
# with open(log_file_path, 'r') as file:
#     for line in file:
#         match = re.search(r"D Loss: (-?\d+\.\d+)", line)
#         if match:
#             d_losses.append(float(match.group(1)))
        
#         match = re.search(r"G Loss: (-?\d+\.\d+(e-?\d+)*)", line)
#         if match:
#             g_losses.append(float(match.group(1)))

t_losses = []

with open(log_file_path, 'r') as file:
    for line in file:
        match = re.search(r"val loss = (-?\d+\.\d+)", line)
        if match:
            t_losses.append(float(match.group(1)))

plt.figure(figsize=(10, 5))
# plt.plot(d_losses, label='Discriminator Loss')
# plt.plot(g_losses, label='Generator Loss')
plt.plot(t_losses, label='Transformer Loss')
plt.title('Transformer Loss during Training')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
